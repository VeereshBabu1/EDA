```python
# Replacing with "₹", ","....
df['discounted_price'] = df['discounted_price'].str.replace("₹", "").str.replace(",", "")
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['actual_price'] = df['actual_price'].str.replace("₹", "").str.replace(",", "")
df['actual_price'] = df['actual_price'].astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace("%", "")
df['discount_percentage'] = df['discount_percentage'].astype(int)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
#df['rating'] = df['rating'].str.replace("|", "").str.replace("", "")
#df['rating'] = df['rating'].astype(float)
```

```python
df['rating_count'] = df['rating_count'].str.replace(",", "")
df['rating_count'] = df['rating_count'].astype(float)
```

6. What are the most popular product keywords?

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from collections import Counter

all_keywords = []
for product_name in df_clean['product_name']:
    # Tokenize the product name
    tokens = word_tokenize(str(product_name))
    # Filter out non-alphabetic tokens (e.g., numbers, punctuation)
    keywords = [token.lower() for token in tokens if token.isalpha()]
    # Add the keywords to the list of all keywords
    all_keywords.extend(keywords)

# Count the frequency of each keyword
keyword_counts = Counter(all_keywords)

# Get the most common keywords
most_common_keywords = keyword_counts.most_common(10)  # Adjust the number to get more or fewer keywords

# Display the most common keywords
print("Most Popular Product Keywords:")
for keyword, count in most_common_keywords:
    print(f"{keyword}: {count}")
```

7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.

```python
top_artists = dp.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(5).index

# Create an empty DataFrame to store the results
most_popular_tracks = pd.DataFrame(columns=['Artist', 'Track Name', 'Popularity'])

# For each top artist, find their most popular track
for artist in top_artists:
    artist_tracks = dp[dp['Artist'] == artist]
    most_popular_track = artist_tracks.loc[artist_tracks['Popularity'].idxmax()]
    most_popular_tracks = most_popular_tracks.append({
        'Artist': most_popular_track['Artist'],
        'Track Name': most_popular_track['Track Name'],
        'Popularity': most_popular_track['Popularity']
    }, ignore_index=True)

# Display the results
print("Most Popular Tracks for the Top 5 Most Popular Artists:")
print(most_popular_tracks)
```

10. How does the distribution of track popularity vary for different artists?
Visualize this using a swarm plot or a violin plot.

```python
top_artists = dp['Artist'].value_counts().head(10).index

filtered_data = dp[dp['Artist'].isin(top_artists)]

plt.figure(figsize=(14, 8))
sns.swarmplot(x='Artist', y='Popularity', data=filtered_data)
plt.title('Distribution of Track Popularity for Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45)
plt.show()
```
